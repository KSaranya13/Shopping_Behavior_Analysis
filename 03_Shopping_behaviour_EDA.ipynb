```python
import matplotlib.pyplot as plt

plt.style.use('default')   # optional for clean theme

# Create 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# ================================
# 1. Age Distribution
axes[0, 0].hist(df['Age'], bins=12, edgecolor='black')
axes[0, 0].set_title("Age Distribution of Customers", fontsize=14)
axes[0, 0].set_xlabel("Age")
axes[0, 0].set_ylabel("Number of Customers")
axes[0, 0].grid(True, linestyle=' ', alpha=0.4)

# ================================
# 2. Purchase Amount Distribution
axes[0, 1].hist(df['Purchase Amount (USD)'], bins=10, edgecolor='black')
axes[0, 1].set_title("Purchase Amount Distribution", fontsize=14)
axes[0, 1].set_xlabel("Purchase Amount (USD)")
axes[0, 1].set_ylabel("Number of Customers")
axes[0, 1].grid(True, linestyle=' ', alpha=0.4)

# ================================
# 3. Review Rating Distribution
axes[1, 0].hist(df['Review Rating'], bins=10, edgecolor='black')
axes[1, 0].set_title("Review Rating Distribution", fontsize=14)
axes[1, 0].set_xlabel("Review Rating")
axes[1, 0].set_ylabel("Number of Customers")
axes[1, 0].grid(True, linestyle=' ', alpha=0.4)

# ================================
# 4. Previous Purchases Distribution
axes[1, 1].hist(df['Previous Purchases'], bins=10, edgecolor='black')
axes[1, 1].set_title("Previous Purchases Distribution", fontsize=14)
axes[1, 1].set_xlabel("Number of Previous Purchases")
axes[1, 1].set_ylabel("Number of Customers")
axes[1, 1].set_style = ("whitegrid")
axes[1, 1].grid(True, linestyle=' ', alpha=0.4)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Box Plot for outlier Detection
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,8))

plt.subplot(2, 2, 1)
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")

plt.subplot(2, 2, 2)
sns.boxplot(x=df['Purchase Amount (USD)'])
plt.title("Boxplot of Purchase Amount (USD)")

plt.subplot(2, 2, 3)
sns.boxplot(x=df['Review Rating'])
plt.title("Boxplot of Review Rating")

plt.subplot(2, 2, 4)
sns.boxplot(x=df['Previous Purchases'])
plt.title("Boxplot of Previous Purchases")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
corr = df[['Age','Purchase Amount (USD)','Review Rating','Previous Purchases']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.show()
```

```python
#Product Category Distribution using Bar Chart
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Category',
    order=df['Category'].value_counts().index,   # sort by highest count
    palette='pastel'
)

plt.title("Product Category Distribution", fontsize=16)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Number of Purchases", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Season Distribution using Bar Chart
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Season',
    order=df['Season'].value_counts().index,   # sort by highest count
    palette='pastel'
)

plt.title("Season-wise Purchase Distribution", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Purchases", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
